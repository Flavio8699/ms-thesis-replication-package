```python
import pandas as pd
```

Build dataframe for 22 degrees

```python
df22_1 = pd.read_csv(
    'datasets/debian/dataset_2024-05-04_21-00_26x_48C_6-conc-22C.csv')[1:]
df22_2 = pd.read_csv(
    'datasets/debian/dataset_2024-05-05_21-00_26x_48C_6-conc-22C.csv')[1:]

df22 = pd.concat([df22_1, df22_2])

for column in df22.filter(like='_avg_watts_consumption').columns:
    prefix = column.split('_avg_watts_consumption')[0]

    duration = df22[f'{prefix}_duration']
    watts = df22[column]

    df22[f'{prefix}_joules_consumption'] = (watts - 13.2) * duration

df22.reset_index(drop=True, inplace=True)
df22
```

```python
df24_1 = pd.read_csv(
    'datasets/debian/dataset_2024-05-08_21-00_26x_48C_6-conc-24C.csv')[1:]
df24_2 = pd.read_csv(
    'datasets/debian/dataset_2024-05-09_21-00_26x_48C_6-conc-24C.csv')[1:]

df24 = pd.concat([df24_1, df24_2])

for column in df24.filter(like='_avg_watts_consumption').columns:
    prefix = column.split('_avg_watts_consumption')[0]

    duration = df24[f'{prefix}_duration']
    watts = df24[column]

    df24[f'{prefix}_joules_consumption'] = (watts - 13.2) * duration

df24.reset_index(drop=True, inplace=True)
df24
```

Mean values for 22 degrees

```python
print(df22['pipeline_duration'].mean())
print(df22['pipeline_joules_consumption'].mean())
print(df22['pipeline_avg_ram_consumption'].mean())
print(df22['pipeline_avg_disk_consumption'].mean())
```

Mean values for 24 degrees

```python
print(df24['pipeline_duration'].mean())
print(df24['pipeline_joules_consumption'].mean())
print(df24['pipeline_avg_ram_consumption'].mean())
print(df24['pipeline_avg_disk_consumption'].mean())
```

Function to calculate percentage change between both temps

```python
def P_diff(m24, m22):
    return abs(m24-m22)/((m24+m22)/2) * 100
```

Percentage change

```python
print(P_diff(df24['pipeline_duration'].mean(), df22['pipeline_duration'].mean()))
print(P_diff(df24['pipeline_joules_consumption'].mean(), df22['pipeline_joules_consumption'].mean()))
print(P_diff(df24['pipeline_avg_ram_consumption'].mean(), df22['pipeline_avg_ram_consumption'].mean()))
print(P_diff(df24['pipeline_avg_disk_consumption'].mean(), df22['pipeline_avg_disk_consumption'].mean()))
```
